```python
from random import randint
```

```python
musteriler = [ "Mus1" , "Mus2" , "Mus3" , "Mus4" , "Mus5" ]
```

```python
harcamalar = {}
```

```python
for i in musteriler:
    harcamalar[i] = randint(0,2000)

print (harcamalar)
```

```python
def filtre( veriList , esik ):
    yeniListe = []
    for i in veriList:
        if( veriList[i] > esik ):
            yeniListe.append(i)
    return yeniListe
            
```

```python
def segmentasyon(veriList):
    stdUye = []
    gumusUye = []
    altinUye = []

    for i in veriList:
        if veriList[i] < 500:
            stdUye.append(i)
        elif veriList[i]>= 500 and veriList[i] < 1500:
            gumusUye.append(i)
        else:
            altinUye.append(i)

    return stdUye , gumusUye , altinUye
```

```python
secim = 0
while(secim != 3):
    secim = int(input("Filtreleme yapmak için 1'e, segmentasyon raporu için 2'ye, çıkmak için 3'e basın: "))
    if secim == 1:
        esikDegeri = int(input("Filtrelemek istediğiniz eşik değerini girin: "))
        sonuc = filtre(harcamalar, esikDegeri)
        print("\n--- Filtreleme Sonucu ---")
        print(sonuc)
        print("-------------------------\n")
        continue
    elif secim == 2:
        st, gm, al = segmentasyon(harcamalar)
        print("\n--- Segmentasyon Raporu ---")
        print("Standart Müşteriler :", st)
        print("Gümüş Müşteriler    :", gm)
        print("Altın Müşteriler     :", al)
        print("---------------------------\n")
        continue
    elif secim == 3:
        print("Çıkış yaptınız.")
        break
    else:
        print("Hatalı seçim yaptınız.")
        continue
```

```python
musteriler2 = [ "Mus1" , "Mus2" , "Mus3" , "Mus4" , "Mus5" ]
```

```python
harcamalar2 = {}
```

```python
for i in musteriler2:
    harcamalar2[i] = randint(0,2000)
print(harcamalar2)
```

```python
def filtreleme(veriList , esik):
    return list(filter(lambda i: veriList[i] > esik , veriList))
```

```python
def segmente(veriList):
    standart = list(filter(lambda i: veriList[i] < 500 , veriList))
    gumus = list(filter(lambda i: veriList[i] >= 500 and veriList[i] < 1500 , veriList))
    altin = list(filter(lambda i: veriList[i] >= 1500 , veriList))
    return standart , gumus , altin
```

```python
secim = 0
while(secim != 3):
    secim = int(input("Filtreleme yapmak için 1'e, segmentasyon raporu için 2'ye, çıkmak için 3'e basın: "))
    if secim == 1:
        esikDegeri = int(input("Filtrelemek istediğiniz eşik değerini girin: "))
        sonuc = filtre(harcamalar, esikDegeri)
        print("\n--- Filtreleme Sonucu ---")
        print(sonuc)
        print("-------------------------\n")
        continue
    elif secim == 2:
        st, gm, al = segmentasyon(harcamalar2)
        
        st_etiketli = list(map(lambda isim: isim + " [Standart]", st))
        gm_etiketli = list(map(lambda isim: isim + " [Gümüş]", gm))
        al_etiketli = list(map(lambda isim: isim + " [Altın]", al))
        
        print("\n--- Etiketli Segmentasyon Raporu ---")
        print("Standartlar :", st_etiketli)
        print("Gümüşler    :", gm_etiketli)
        print("Altınlar     :", al_etiketli)
        print("-------------------------------------\n")
        continue
    elif secim == 3:
        print("Çıkış yaptınız.")
        break
    else:
        print("Hatalı seçim yaptınız.")
        continue
```
